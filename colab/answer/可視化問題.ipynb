## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
data = pd.read_csv('data/hist_data.csv')

data_a = data.iloc[:, 0]
data_b = data.iloc[:, 1]

plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')

plt.hist(data_a, bins=30, color='y', alpha=0.5, label='aデータ')
plt.hist(data_b, bins=30, color='c', alpha=0.5, label='bデータ')

plt.legend()
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
data = pd.read_csv('data/scatter_data.csv')

data_length = data.iloc[:, 0]
data_width = data.iloc[:, 1]
data_class = data.iloc[:, 2]

plt.title('クラス０～２のlengthとwidthに関する散布図')
plt.xlabel(data.keys()[0])
plt.ylabel(data.keys()[1])

plt.scatter(data_length[data_class==0], data_width[data_class==0], color='r', label='class0')
plt.scatter(data_length[data_class==1], data_width[data_class==1], color='b', label='class1')
plt.scatter(data_length[data_class==2], data_width[data_class==2], color='g', label='class2')

plt.grid(True)
plt.legend()
plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
data = pd.read_csv('data/line_data.csv')

data_period = data.iloc[:, 0]
data_count = data.iloc[:, 1]
data_cost = data.iloc[:, 2]

plt.plot(data_period, data_count, marker='o')
plt.xticks(rotation=45)

plt.title('期間ごとの件数の遷移')
plt.xlabel(data.keys()[0])
plt.ylabel(data.keys()[1])

plt.grid(True)
plt.show()
```

```python
plt.bar(data_period, data_cost)
plt.xticks(rotation=45)

plt.title('期間ごとの費用の遷移')
plt.xlabel(data.keys()[0])
plt.ylabel(data.keys()[2])

plt.grid(True)
plt.show()
```
